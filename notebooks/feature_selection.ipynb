```python
# Feature Selection
#This section performs feature selection on the combined dataset to reduce dimensionality and improve model performance.
```

```python
import numpy as np
import pandas as pd

combined_data = np.load("../data/processed/combined_data.npy")
combined_labels = np.load("../data/processed/combined_labels.npy")

# Load combined dataset created after preprocessing
```

```python
print(combined_data.shape)
print(len(combined_labels))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_non_negative = scaler.fit_transform(combined_data)

# Convert all values to range [0,1]
# Required for Chi-square test
```

```python
print(np.min(data_non_negative), np.max(data_non_negative))
```

```python
from sklearn.feature_selection import SelectKBest, chi2

selector_2000 = SelectKBest(score_func=chi2, k=2000)
data_chi2_2000 = selector_2000.fit_transform(data_non_negative, combined_labels)

# select top 2000 features
```

```python
print(data_chi2_2000.shape)
```

```python
selector_1500 = SelectKBest(score_func=chi2, k=1500)
data_chi2_1500 = selector_1500.fit_transform(data_non_negative, combined_labels)

# select top 1500 features
```

```python
print(data_chi2_1500.shape)
```

```python
selector_1000 = SelectKBest(score_func=chi2, k=1000)
data_chi2_1000 = selector_1000.fit_transform(data_non_negative, combined_labels)

# select top 1000 features
```

```python
print(data_chi2_1000.shape)
```

```python
np.save("../data/processed/chi2_2000.npy", data_chi2_2000)
np.save("../data/processed/chi2_1500.npy", data_chi2_1500)
np.save("../data/processed/chi2_1000.npy", data_chi2_1000)

# save chi-square feature sets
```

```python
import os
print(os.listdir("../data/processed"))
```

```python
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(combined_data, combined_labels)

# compute MI score for each feature
```

```python
print(mi_scores.shape)
print(mi_scores[:10])
print(np.isnan(mi_scores).sum())
```

```python
mi_indices = np.argsort(mi_scores)[::-1]

# sort features by importance (descending)
```

```python
print(mi_indices[:10])
```

```python
mi_2000 = combined_data[:, mi_indices[:2000]]

# top 2000 features
```

```python
mi_1500 = combined_data[:, mi_indices[:1500]]

# top 1500 features
```

```python
mi_1500 = combined_data[:, mi_indices[:1500]]

# top 1500 features
```

```python
print(mi_2000.shape)
print(mi_1500.shape)
print(mi_1000.shape)
```

```python
mi_1000 = combined_data[:, mi_indices[:1000]]

# top 1000 features
```

```python
print(mi_2000.shape)
print(mi_1500.shape)
print(mi_1000.shape)
```

```python
np.save("../data/processed/mi_2000.npy", mi_2000)
np.save("../data/processed/mi_1500.npy", mi_1500)
np.save("../data/processed/mi_1000.npy", mi_1000)

# save MI feature sets
```
